```python
#2012 election

#open the site
with open("2012 United States presidential election - Wikipedia.html") as file_reader:
    soup = BeautifulSoup(file_reader, "lxml")

#go through all the text and all votes shares that favored republican
state_12_rep = [] 
obama_12_rep = []
romney_12_rep = []

text_all_rep = soup.find_all("tr", {"style":"background-color:#FFB6B6"})
#print(text_all)
for i in range (2, 26):
    rate_all_rep = text_all_rep[i].get_text()
    state_rep = re.findall("[a-zA-Z]+", rate_all_rep)
    state_12_rep.append(state_rep)
    obama_rate_rep = re. findall("\S*%", rate_all_rep)[0]
    obama_12_rep.append(obama_rate_rep)
    romney_rate_rep = re. findall("\S*%", rate_all_rep)[1]
    romney_12_rep.append(romney_rate_rep)

rep_rate_2012_col = {'State': state_12_rep, 'Obama_rate': obama_12_rep, 'Romney_rate': romney_12_rep}
rep_rate_2012 = pd.DataFrame(data=rep_rate_2012_col)

#go through all the text and all votes shares that favored democrat

state_12_dem = [] 
obama_12_dem = []
romney_12_dem = []

text_all_dem = soup.find_all("tr", {"style":"background-color:#B0CEFF"})
#print(text_all)
for i in range (2, 29):
    rate_all_dem = text_all_dem[i].get_text()
    state_dem = re.findall("[a-zA-Z]+", rate_all_dem)
    state_12_dem.append(state_dem)
    obama_rate_dem = re. findall("\S*%", rate_all_dem)[0]
    obama_12_dem.append(obama_rate_dem)
    romney_rate_dem = re. findall("\S*%", rate_all_dem)[1]
    romney_12_dem.append(romney_rate_dem)

dem_rate_2012_col = {"State":state_12_dem, 'Obama_rate': obama_12_dem, 'Romney_rate': romney_12_dem}
dem_rate_2012 = pd.DataFrame(data=dem_rate_2012_col)

rate_2012_result = pd.concat([dem_rate_2012, rep_rate_2012])
print("2012 results")

#link vote share to state
rate_2012_result = rate_2012_result.reset_index()
rate_2012_result = rate_2012_result.drop('index', axis = 1)
#rate_2012_result
state_abbrev = ['']*len(rate_2012_result)

state = rate_2012_result['State']

for i in range(len(state)):
    for j in range(len(state[i])):
        for k in range(len(states_abbrev)):
            if state[i][j] == states_abbrev[k]:
                state_abbrev[i] = state[i][j]
rate_2012_result['State'] = state_abbrev

rate_2012_result_2 = rate_2012_result.copy()

#fill out dataframe
rate_2012_result['candidate'] = "obama_12"
rate_2012_result = rate_2012_result.drop('Romney_rate', axis=1)
rate_2012_result["Vote Share"] = rate_2012_result['Obama_rate']
rate_2012_result = rate_2012_result.drop('Obama_rate', axis=1)
rate_2012_result


rate_2012_result_2['candidate'] = "romney_12"
rate_2012_result_2 = rate_2012_result_2.drop('Obama_rate', axis=1)
rate_2012_result_2["Vote Share"] = rate_2012_result_2['Romney_rate']
rate_2012_result_2 = rate_2012_result_2.drop('Romney_rate', axis=1)
rate_2012_result_2



rate_2012_final = pd.concat([rate_2012_result, rate_2012_result_2])
sort = True
rate_2012_final
```
