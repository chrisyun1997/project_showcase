```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#load csv files
spending_sum = pd.read_csv('spending.csv')
voting = pd.read_csv('voting.csv')
```

```python
#clean and merge datasets
voting = voting.drop('Unnamed: 0', axis = 1)
spending_sum = spending_sum.drop('Unnamed: 0', axis = 1)
spending_sum['State'] = spending_sum['recipient_state']
vote_spend = pd.merge(spending_sum,voting,on = ['candidate', 'State'])
```

```python
#relabel variables and sort variables
vote_spend['Year'] = vote_spend['election_year']
vote_spend['amount_spent'] = vote_spend['disbursement_amount']
vote_spend['state'] = vote_spend['State']
vote_spend['vote_share'] = vote_spend['Vote Share']
vote_spend['year']= vote_spend['Year']
vote_spend = vote_spend[['year','candidate','state','amount_spent','vote_share']]
```

```python
#remove % and turn vote_share into float
vote_share = vote_spend['vote_share']
for i in range(len(vote_share)):
    if "%" in vote_share[i]:
        vote_share[i] = vote_share[i].replace("%","")
    vote_share[i] = float(vote_share[i])
```

```python
#Add a column party that indicates a candidate's party
party = [""]*len(vote_spend)
vote_spend['party'] = party
for i in range(len(vote_spend)):
    if vote_spend['candidate'][i] == "reagan_80":
        vote_spend['party'][i] = "Republican"
    elif vote_spend['candidate'][i] == "carter_80":
        vote_spend['party'][i] = "Democratic"
    elif vote_spend['candidate'][i] == "anderson_80":
        vote_spend['party'][i] = "Independent"
    elif vote_spend['candidate'][i] == "clark_80":
        vote_spend['party'][i] = "Libertarian"
    elif vote_spend['candidate'][i] == "reagan_84":
        vote_spend['party'][i] = "Republican"
    elif vote_spend['candidate'][i] == "mondale_84":
        vote_spend['party'][i] = "Democratic"
    elif vote_spend['candidate'][i] == "bush_88":
        vote_spend['party'][i] = "Republican"
    elif vote_spend['candidate'][i] == "dukakis_88":
        vote_spend['party'][i] = "Democratic"
    elif vote_spend['candidate'][i] == "bush_92":
        vote_spend['party'][i] = "Republican"
    elif vote_spend['candidate'][i] == "clinton_92":
        vote_spend['party'][i] = "Democratic"
    elif vote_spend['candidate'][i] == "perot_92":
        vote_spend['party'][i] = "Independent"
    elif vote_spend['candidate'][i] == "dole_96":
        vote_spend['party'][i] = "Republican"
    elif vote_spend['candidate'][i] == "clinton_96":
        vote_spend['party'][i] = "Democratic"
    elif vote_spend['candidate'][i] == "perot_96":
        vote_spend['party'][i] = "Reform"
    elif vote_spend['candidate'][i] == "bush_00":
        vote_spend['party'][i] = "Republican"
    elif vote_spend['candidate'][i] == "gore_00":
        vote_spend['party'][i] = "Democratic"
    elif vote_spend['candidate'][i] == "nader_00":
        vote_spend['party'][i] = "Green"
    elif vote_spend['candidate'][i] == "bush_04":
        vote_spend['party'][i] = "Republican"
    elif vote_spend['candidate'][i] == "kerry_04":
        vote_spend['party'][i] = "Democratic"
    elif vote_spend['candidate'][i] == "mccain_08":
        vote_spend['party'][i] = "Republican"
    elif vote_spend['candidate'][i] == "obama_08":
        vote_spend['party'][i] = "Democratic"
    elif vote_spend['candidate'][i] == "romney_12":
        vote_spend['party'][i] = "Republican"
    elif vote_spend['candidate'][i] == "obama_12":
        vote_spend['party'][i] = "Democratic"
    elif vote_spend['candidate'][i] == "clinton_16":
        vote_spend['party'][i] = "Democratic"
    elif vote_spend['candidate'][i] == "trump_16":
        vote_spend['party'][i] = "Republican"
    elif vote_spend['candidate'][i] == "johnson_16":
        vote_spend['party'][i] = "Libertarian"
    elif vote_spend['candidate'][i] == "stein_16":
        vote_spend['party'][i] = "Green"
    else:
        print("Oops!")
```

```python
#add a cost per vote share column
cost_per_share = []
for i in range(len(vote_spend)):
    cost_per_share.append(vote_spend['amount_spent'][i]/vote_spend['vote_share'][i])
vote_spend['cost_per_vote_share'] = cost_per_share
```

```python
#partition parties into three kinds: dem, rep, and third
party_partition = []
for party in vote_spend['party']:
    if party == "Democratic":
        party_partition.append(party)
    elif  party == "Republican":
        party_partition.append(party)
    else:
        party_partition.append("Third")
vote_spend['party_partition'] = party_partition
```

```python
#only use data where spending is positive in a state
is_spending = vote_spend[vote_spend['amount_spent'] > 0]
is_spending.reset_index(drop = True)
```

```python
#add dummies for candidate's party affiliation
is_spending_dummy = is_spending 
party_dummy = []
for i in is_spending_dummy['party_partition']: 
    if i == "Republican": 
        party_dummy.append(1)
    if i == "Third": 
        party_dummy.append(2)
    if i == "Democratic": 
        party_dummy.append(0)

#add party_dummy to is_spending
is_spending_dummy.loc[:,"party_dummy"] = party_dummy
is_spending_dummy.head()
```
